```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read sales_data.csv
df = pd.read_csv('sales_data.csv')
print(df.head())
```

```python
# 1.get the sum of the column revenue in the data frame
total_revenue = df['revenue'].sum()
print(f"total_revenue: {total_revenue}")
```

```python
# 2.Which product had the highest revenue? How much revenue did it generate?
# get the product name with the highest revenue
index = df['revenue'].idxmax()
product_name_max_revenue = df['product'][index]
max_revenue = df['revenue'][index]
print(f"product_name: {product_name_max_revenue}, max_revenue: {max_revenue}")
```

```python
df_smartphone = df[df['product'] == product_name_max_revenue]
# get the total quantity sold of this product
total_quantity_smartphone = df_smartphone['quantity'].sum()
print(f"total_quantity_smartphone: {total_quantity_smartphone}")
# get the average revenue of this product
average_revenue_smartphone = df_smartphone['revenue'].mean()
print(f"average_revenue_smartphone: {average_revenue_smartphone}")
```

```python
# average price of all products in the company
average_price_of_all_products = df['price'].mean()
print(f"average_price: {average_price_of_all_products}")
# average revenue of all products in the company
average_revenue_of_all_products = df['revenue'].mean()
print(f"average_revenue: {average_revenue_of_all_products}")
```

```python
# getting the average price of each product individually
# get the average price of each product individually
average_price_of_each_product = df.groupby('product')['price'].mean()
print(f"average_price_of_each_product: {average_price_of_each_product}")
# draw a bar chart to show the average price of each product individually
average_price_of_each_product.plot.bar()
```

```python
# getting the total quantity of products sold by the company individually
# get the total quantity of products sold by the company individually
total_quantity_of_each_product = df.groupby('product')['quantity'].sum()
print(f"total_quantity_of_each_product: {total_quantity_of_each_product}")
# draw a bar chart to show the total quantity of products sold by the company individually
total_quantity_of_each_product.plot.bar()
```

```python
# Which category had the highest revenue? How much revenue did it generate?
# get the category name with the highest revenue
index = df['revenue'].idxmax()
category_name_max_revenue = df['category'][index]
max_revenue = df['revenue'][index]
print(f"category_name: {category_name_max_revenue}, max_revenue: {max_revenue}")
```

```python
# getting the average revenue of each category individually
# get the average revenue of each category individually
average_revenue_of_each_category = df.groupby('category')['revenue'].mean()
print(f"average_revenue_of_each_category: {average_revenue_of_each_category}")
# draw a pie chart to show the average revenue of each category individually with precetages to the pie chart
# add legend to the pie chart to show the category name and the percentage of the revenue
average_revenue_of_each_category.plot.pie(autopct='%1.1f%%')
# edit the legend to be outside the pie chart
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
# make the pie chart a perfect circle
plt.axis('equal')
# show the pie chart
plt.show()
```

```python
# getting the average revenue of each product individually
# get the average revenue of each product individually
average_revenue_of_each_product = df.groupby('product')['revenue'].mean()
print(f"average_revenue_of_each_product: {average_revenue_of_each_product}")
# draw a pie chart to show the average revenue of each product individually with precetages to the pie chart
# add legend to the pie chart to show the product name and the percentage of the revenue
average_revenue_of_each_product.plot.pie(autopct='%1.1f%%')
# edit the legend to be outside the pie chart
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
# make the pie chart a perfect circle
plt.axis('equal')
# show the pie chart
plt.show()
```

```python
# take copy form the data frame
df_copy = df.copy()
# add a new column to the data frame called month
df_copy['month'] = df_copy['date'].str[5:7]
# get the average revenue of each month
average_revenue_of_each_month = df_copy.groupby('month')['revenue'].mean()
print(f"average_revenue_of_each_month: {average_revenue_of_each_month}")
# draw a bar chart to show the average revenue of each month
average_revenue_of_each_month.plot.bar()
# add title to the bar chart
plt.title('average revenue of each month')
# add x label to the bar chart
plt.xlabel('month')
# add y label to the bar chart
plt.ylabel('average revenue in $')
# show the bar chart
plt.show()
```

```python
# get the month name with the highest revenue
index = average_revenue_of_each_month.idxmax()
month_name_max_revenue = index
max_revenue = average_revenue_of_each_month[index]
print(f"month_name: {month_name_max_revenue}, max_revenue: {max_revenue}")
```

```python
# get all products sold in the each quarter of the year
df_q1 = df_copy[(df_copy['month'] >= '01') & (df_copy['month'] <= '03')]
df_q2 = df_copy[(df_copy['month'] >= '04') & (df_copy['month'] <= '06')]
df_q3 = df_copy[(df_copy['month'] >= '07') & (df_copy['month'] <= '09')]
df_q4 = df_copy[(df_copy['month'] >= '10') & (df_copy['month'] <= '12')]
# get the total revenue of the each quarter of the year
total_revenue_q1 = df_q1['revenue'].sum()
total_revenue_q2 = df_q2['revenue'].sum()
total_revenue_q3 = df_q3['revenue'].sum()
total_revenue_q4 = df_q4['revenue'].sum()
print(f"total_revenue_q1: {total_revenue_q1}")
print(f"total_revenue_q2: {total_revenue_q2}")
print(f"total_revenue_q3: {total_revenue_q3}")
print(f"total_revenue_q4: {total_revenue_q4}")
# get the quarter with the highest revenue
revenue_of_the_year = [total_revenue_q1, total_revenue_q2, total_revenue_q3, total_revenue_q4]
quarter_name = ['Q1', 'Q2', 'Q3', 'Q4']
index = revenue_of_the_year.index(max(revenue_of_the_year))
quarter_name_max_revenue = quarter_name[index]
max_revenue = revenue_of_the_year[index]
print(f"quarter_name of max revenue: {quarter_name_max_revenue}, max_revenue: {max_revenue}")
# draw a bar chart to show the total revenue of the each quarter of the year
plt.bar(quarter_name, revenue_of_the_year)
# add title to the bar chart
plt.title('total revenue of the each quarter of the year')
# add x label to the bar chart
plt.xlabel('quarter')
# add y label to the bar chart
plt.ylabel('total revenue in $')
# show the bar chart
plt.show()
```
